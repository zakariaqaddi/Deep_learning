<h2>Importing librairies</h2>

```python
import pandas as pd 
import keras
from keras.models import Sequential
from keras.layers import Dense
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from statistics import stdev 
```

```python
def average (a) : 
    return sum(a)/len(a)
```

<h2>Importing data</h2>

```python
URL='https://cocl.us/concrete_data' #data URL
data=pd.read_csv(URL) #Reading data
#data.head() #printing the first 5 rows
```

```python
data_columns=data.columns  
predectors=data[data_columns[data_columns !='Strength']] #Predectors
target=data['Strength'] #Target
```

```python
cons=predectors.shape[1]
```

<h2>A Regression Function  </h2>

```python
def regressionmodel () : 
    model=Sequential() #create the model
    model.add(Dense(10,activation='relu',input_shape=(cons,))) #add the first hidden layer with 10 nodes 
    model.add(Dense(1)) #dense the modul in one output 
    model.compile(optimizer='adam',loss='mean_squared_error') #compile the model using the adam optimizer
    return model 
```

```python
model=regressionmodel () #call the regression function (model)
```

```python
Mean_square_error_a=[] #Mean_square_error list
```

```python
for i in range (49):
    X=predectors
    y=target
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
    model.fit(X_train, y_train,epochs=50) #train the model 
    y_predict=model.predict(X_test)
    Mean_square_error_a.append( mean_squared_error(y_test,y_predict))
print('the mean of the mean square error A : {} the std of the mean square error is A : {}'
      .format(average(Mean_square_error_a),stdev(Mean_square_error_a)))
```

<h2>B Normalize data</h2>

```python
Mean_square_error_b=[]
```

```python
modelb=regressionmodel()
predectors=(predectors-predectors.mean())/(predectors.std())
for i in range (49):
    X=predectors
    y=target
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
    modelb.fit(X_train, y_train,epochs=50) #train the model 
    y_predict=modelb.predict(X_test)
    Mean_square_error_b.append( mean_squared_error(y_test,y_predict))
print('the mean of the mean square error B : {} the std of the mean square error B is : {}'
      .format(average(Mean_square_error_b),stdev(Mean_square_error_b)))
```

```python
if average(Mean_square_error_a)>average(Mean_square_error_b) : 
    print('the model in A is efficace than the model B')
else:
     print('the model in B is efficace than the model A')
```

<h2>D //Increase the number of hidden layers</h2> 

```python
def regressionmodelD () : 
    model=Sequential() #create the model
    model.add(Dense(10,activation='relu',input_shape=(cons,))) #add the first hidden layer with 10 nodes 
    model.add(Dense(10,activation='relu'))
    model.add(Dense(10,activation='relu'))
    model.add(Dense(1)) #dense the modul in one output 
    model.compile(optimizer='adam',loss='mean_squared_error') #compile the model using the adam optimizer
    return model 
```

```python
modeld=regressionmodelD()
```

```python
Mean_square_error_d=[]
```

```python
modeld=regressionmodelD()
predectors=(predectors-predectors.mean())/(predectors.std())
for i in range (49):
    X=predectors
    y=target
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
    modeld.fit(X_train, y_train,epochs=50) #train the model 
    y_predict=modeld.predict(X_test)
    Mean_square_error_d.append( mean_squared_error(y_test,y_predict))
print('the mean of the mean square error d : {} the std of the mean square error d is : {}'
      .format(average(Mean_square_error_d),stdev(Mean_square_error_d)))
```

```python
if average(Mean_square_error_c)>average(Mean_square_error_d) : 
    print('the model in d is efficace than the model c')
else:
     print('the model in c is efficace than the model d')
```
